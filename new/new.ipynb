```python
import pandas as pd
import numpy as np
data =  list(np.random.randn(10000))
print(pd.Series(data).skew())
print(pd.Series(data).kurt())
```

```python
# 计算第一阶矩（均值）
mean = np.mean(data)
print("第一阶矩（均值）：", mean)

# 计算前四阶中心矩
moments = np.zeros(4)
for i in range(1, 5):
    moments[i-1] = np.mean((data - mean)**i)
print("前四阶中心矩：", moments)
```

```python
# 计算第一阶矩（均值）
mean_value = np.mean(data)

# 计算前四阶中心矩
central_moment_2 =np.mean((data - mean_value)**2)
central_moment_3 = np.mean((data - mean_value)**3)
central_moment_4 = np.mean((data - mean_value)**4)

print(f"Mean (First Moment): {mean_value}")
print(f"Central Moment 2: {central_moment_2}")
print(f"Central Moment 3: {central_moment_3}")
print(f"Central Moment 4: {central_moment_4}")
```

```python
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt
import seaborn as sns
                                                
data =  list(np.random.randn(10000)) # 满足高斯分布的10000个数
sns.distplot(data, fit=norm)
(mu, sigma) = norm.fit(data)
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],loc='best')
plt.ylabel('Frequency')
plt.title('Distribution')
```
